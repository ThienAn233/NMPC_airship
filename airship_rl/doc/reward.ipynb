```python
import matplotlib.pyplot as plt
import numpy as np
n   = 10
xs  = np.linspace(0,10,n)
ys  = np.linspace(0,10,n)
zs  = np.linspace(0,10,n)
X,Y,Z = np.meshgrid(xs,ys,zs)
```

```python

def reward(location,target):
    xm, ym, zm = location
    xt, yt, zt = target
    l = np.linalg.norm(np.linalg.cross(location,target))/np.linalg.norm(target)
    da = np.linalg.norm(location-target)
    const = np.linalg.norm(target)
    return -(l+da-const)**2
```

```python
target = np.array([1,1,1])
ws = []
for x,y,z in zip(X.flatten(),Y.flatten(),Z.flatten()):
    ws.append(reward([x,y,z],target))
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.scatter(X.flatten(),Y.flatten(),Z.flatten(),c=ws)
plt.colorbar(ax.scatter(X.flatten(),Y.flatten(),Z.flatten(),c=ws))
```

```python
# 2D case
n   = 10
xs  = np.linspace(0,2,n)
ys  = np.linspace(0,2,n)
X,Y = np.meshgrid(xs,ys)
def reward_2d(location,target):
    xm, ym = location
    xt, yt = target
    l = np.linalg.norm(np.cross(location,target))/np.linalg.norm(target)
    da = np.linalg.norm(location-target)
    const = np.linalg.norm(target)
    return -10*np.abs(l+da-const)
target = np.array([1,1])
ws = [] 
for x,y in zip(X.flatten(),Y.flatten()):
    ws.append(reward_2d(np.array([x,y]),target))
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
surf = ax.plot_surface(X, Y, np.array(ws).reshape(n,n), cmap='viridis')
ax.set_xlim(-0,2)
ax.set_ylim(-0,2)
ax.scatter3D(target[0],target[1],10,c='r')
# plt.colorbar(surf)
```
